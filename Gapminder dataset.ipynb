## Introduction

This report is an exploration of data on countries around the world. Gapminder World has collected a variety of data on how people live their lives across the globe. This report will try to identify trends in the following datasets for the United States:
* Life expectancy - The average number of years a newborn child would live if current mortality rates stayed the same
* Happiness score (WHR) - Measured using the Cantril life ladder which is explained below 
* The income per person- GDP/capita with inflation adjusted to 2011 Purchasing Power Parity (PPP)
* The unemployment rate for individuals 15 years and over 
* Working hours per week - Calucualted by the total amount of yearly working hours divided by 52 weeks

#### Cantril Life Ladder Method 
It is the national average response to the question of life evaluations asking the following question: "Please imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst life for you. On which stepof the ladder would you say you personally feel you stand at this time?" The dataset expresses the answers as a percentage

### The questions to explore are as follows:
1. Does a higher income lead to a higher life expectancy?
2. Does a higher unemployment rate lead to a lower life expectancy?
3. Does a higher hours worked per week lead to a higher life expectancy?
4. Does a higher hours worked per week lead to a happier nation?
5. Does a higher income mean a nation is more happy?
6. Does a higher unemployment rate mean a nation is less happy?

The charts created show a weak positive correlation but also show that towards the right side a little above 78 years old we see unemployment numbers of all ranges. The average, combined genders, US life expectancy in 2017, which is the range for the data used was 78.6 according to the [CDC](https://www.cdc.gov/nchs/products/databriefs/db328.htm). #### Does a higher hours worked per week lead to a higher life expectancy?
During the late 1990's we can see a significant drop in working hours. And after that change happened, life expectancy did go go up. However we cannot say for certain that this increase in life expectancy is due to shortened work hours. It would be interesting to look further into but this is beyond the scope of this study.

#### Does a higher hours worked per week lead to a happier nation?
No correlation is observed

#### Does a higher income mean a nation is more happy?
We can see a slighlty negative correlation. This implies that a slightly lower income means more happiness. However we cannot say for certain that higher happiness score is a direct causation lower income. There could be many other factors that go into this such as more income could mean more hours worked and therefore less time for leisure and other factors affecting happiness score.

It is worth noting again that the income levels shown in the data is well above the United States poverty line by at least a facor of 3. The federal poverty line for the United States in 2020 is $ 12,760. The poverty line data is sourced from [here](https://www.payingforseniorcare.com/federal-poverty-level) 

#### Does a higher unemployment rate mean a nation is less happy?
No strong correlation is observed between unemployment and the happiness score
